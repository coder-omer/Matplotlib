```python
x = np.arange(0,10)
y = x**2
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('TITLE')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('LARGER')
axes2.set_title('SMALLER')
```

```python
from matplotlib import style
```

```python
style.use('grayscale')

x=np.arange(0,30,0.2)
y=np.sin(x)

plt.grid(True)

plt.plot(x,y)
plt.show()
```

```python
# style.use('grayscale')
style.use('ggplot')

x=np.arange(0,30,0.2)
y1=np.sin(x)
y2=np.cos(x)

plt.title("Sine Function")
# plt.suptitle("I am bigger")

plt.xlabel("Weight of Students")
plt.ylabel("Height of Students")

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```
