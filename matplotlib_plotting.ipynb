```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90, 136, 89, 27]
profit = [40,2,34,12]
```

```python
plt.yticks(ypos,company)
plt.title("US Tech Stocks")

plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g', histtype='step')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
         label=['men','women'],orientation='horizontal')

plt.legend()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels =  ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"] 
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels);
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=1.5, autopct="%0.2f%%",shadow=True, explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=1.5, autopct="%0.2f%%",shadow=True, explode=[0,0.2,0,0,0])
plt.show()
```
